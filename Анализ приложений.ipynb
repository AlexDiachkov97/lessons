```python
unique_playstore = df.drop_duplicates(subset=['App']).reset_index(drop=True)
```

```python
unique_playstore
```

```python
playstore = unique_playstore.copy()
```

```python
for i in range(0, playstore.shape[1]):
    name = playstore.columns[i]
    new_name = name.lower().replace(' ', '_')
    playstore = playstore.rename(columns = {name: new_name})
```

```python
playstore
```

```python
edication_playstore = playstore[(playstore['category'] == 'EDUCATION') & (playstore['reviews'] > 1000)].reset_index(drop=True)
edication_playstore
```

```python
playstore['last_updated'] = pd.to_datetime(playstore['last_updated'])
playstore
```

```python
playstore['price'] = playstore['price'].apply(lambda x: x.replace('$',''))
playstore['price'] = playstore['price'].astype(float)
```

```python
playstore['price'].unique()
```

```python
price_median = round(playstore['price'].median(), 2)
price_mean = round(playstore['price'].mean(), 2)
```
